## Compute Subjectivity

```python
# Function to obtain Subjectivity (means more opinion or emotion based sentiment)
from textblob import TextBlob
def getSub(text):
    return TextBlob(text).sentiment.subjectivity
```

## Compute Polarity

```python
# Function to obtain Polarity (which tells how positive or negative the sentiment of text is)

def getPol(text):
    return TextBlob(text).sentiment.polarity
```

```python
# Apply on dataframe and store in two new columns

requests_df["Subjectivity"] = requests_df["query"].apply(getSub)
requests_df["Polarity"] = requests_df["query"].apply(getPol)
```

## Cleaning the data

```python
# Cleaning the text by using regular expressions
import re
def clean(text):
    text = re.sub(r'@[A-Za-z0-9]+','',text) # Removing @mentions by substituting them with empty string
    text = re.sub(r'#', '', text) # Removing hashtags
    text = re.sub(r'RT[\s]+', '', text) # Removing Retweets
    text = re.sub(r'https?;\/\/\S+', '', text) # Removing hyperlinks
    
    return text

# Applying above function on dataframe
df["content"] = df["content"].apply(clean)

# Printing cleaned dataframe
df
```

```python
# Keeping a copy
emotion_data = df.copy()

# Making all letters lowercase
emotion_data['content'] = emotion_data['content'].apply(lambda x: " ".join(x.lower() for x in x.split()))

# Removing Punctuation, Symbols
emotion_data['content'] = emotion_data['content'].str.replace('[^\w\s]',' ')

emotion_data
```

### Now removing stop words (like 'the', 'an', 'a') using nltk toolkit

```python
import numpy as np
from nltk.corpus import stopwords
from sklearn import preprocessing
```

```python
# Removing Stop Words using NLTK
stop = stopwords.words('english')
emotion_data['content'] = emotion_data['content'].apply(lambda x: " ".join(x for x in x.split() if x not in stop))
```

### Lemmatization and handling repetitions

```python
#Lemmatisation
from textblob import Word
emotion_data['content'] = emotion_data['content'].apply(lambda x: " ".join([Word(word).lemmatize() for word in x.split()]))
#Correcting Letter Repetitions

def de_repeat(text):
    pattern = re.compile(r"(.)\1{2,}")
    return pattern.sub(r"\1\1", text)

emotion_data['content'] = emotion_data['content'].apply(lambda x: " ".join(de_repeat(x) for x in x.split()))

# Code to find the top 10,000 rarest words appearing in the data
freq = pd.Series(' '.join(emotion_data['content']).split()).value_counts()[-10000:]

# Removing all those rarely appearing words from the data
freq = list(freq.index)
emotion_data['content'] = emotion_data['content'].apply(lambda x: " ".join(x for x in x.split() if x not in freq))
```

# Modelling

```python
from sklearn.feature_extraction.text import CountVectorizer
from sklearn import metrics

# Create our vectorizer
vectorizer = CountVectorizer()

# Get the training vectors
vectors = vectorizer.fit_transform(emotion_data["content"])
print(vectors)
```

## 1. Multinomial Naive Bayes

```python
from sklearn.naive_bayes import MultinomialNB


# Build the classifier
mnb = MultinomialNB(alpha=.01)

#  Train the classifier
mnb.fit(vectors, emotion_data["emotion"])

# Get the test vectors
vectors_test = vectorizer.transform(emotion_data["content"])

# Predict and score the vectors
pred = mnb.predict(vectors_test)
acc_score = metrics.accuracy_score(emotion_data["emotion"], pred)
f1_score = metrics.f1_score(emotion_data["emotion"], pred, average='macro')

print('Total accuracy classification score: {}'.format(acc_score))
print('Total F1 classification score: {}'.format(f1_score))
print(pred)
```

```python
from sklearn.metrics import confusion_matrix
cf_matrix = confusion_matrix(emotion_data['emotion'], pred)

# Plotting heatmap
import seaborn as sns
fig, ax = plt.subplots(figsize=(15,8)) 
sns.heatmap(cf_matrix, annot=True, ax=ax)

ax.set_title('Heatmap for Multinomial Naive Bayes', fontdict = {'fontsize' : 20})
plt.show()
```

```python
# Plotting heatmap by percentages
import seaborn as sns
fig, ax = plt.subplots(figsize=(15,8)) 
sns.heatmap(cf_matrix/np.sum(cf_matrix)*100, annot=True, ax=ax)

ax.set_title('Heatmap for Multinomial Naive Bayes', fontdict = {'fontsize' : 20})
plt.show()
```

```python
from sklearn.metrics import classification_report
# Classification report 
print(classification_report(emotion_data["emotion"], pred))
```

### Accuracy of Multinomial Naive Bayes is 73.8%

## 2. Logistic Regression

```python
from sklearn.linear_model import LogisticRegression
# Build the classifier
lr = LogisticRegression()

#  Train the classifier
lr.fit(vectors, emotion_data["emotion"])

# Get the test vectors
vectors_test = vectorizer.transform(emotion_data["content"])

# Predict and score the vectors
pred = lr.predict(vectors_test)
acc_score = metrics.accuracy_score(emotion_data["emotion"], pred)
f1_score = metrics.f1_score(emotion_data["emotion"], pred, average='macro')

print('Total accuracy classification score: {}'.format(acc_score))
print('Total F1 classification score: {}'.format(f1_score))

print((pred))
```

```python
from sklearn.metrics import confusion_matrix
cf_matrix = confusion_matrix(emotion_data['emotion'], pred)

# Plotting heatmap
import seaborn as sns
fig, ax = plt.subplots(figsize=(15,8)) 
sns.heatmap(cf_matrix, annot=True, ax=ax)

ax.set_title('Heatmap for Logistic Regression', fontdict = {'fontsize' : 20})
plt.show()
```

## 3. K Nearest Neighbors

```python
from sklearn.neighbors import KNeighborsClassifier
# Build the classifier
knn = KNeighborsClassifier()

#  Train the classifier
knn.fit(vectors, emotion_data["emotion"])

# Get the test vectors
vectors_test = vectorizer.transform(emotion_data["content"])

# Predict and score the vectors
pred = knn.predict(vectors_test)
acc_score = metrics.accuracy_score(emotion_data["emotion"], pred)
f1_score = metrics.f1_score(emotion_data["emotion"], pred, average='macro')

print('Total accuracy classification score: {}'.format(acc_score))
print('Total F1 classification score: {}'.format(f1_score))

print((pred))
```
